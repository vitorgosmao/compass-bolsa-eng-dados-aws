#### Exercício 2

Leia o arquivo csv googleplaystore.csv e realize as seguintes atividades sobre o dataset utilizando Pandas:

• Faça um gráfico de barras contendo os top 5 apps por número de instalação.

• Faça um gráfico de pizza (pie chart) mostrando as categorias de apps existentes no dataset de acordo com a frequência em que 
elas aparecem.

• Mostre qual o app mais caro existente no dataset.

• Mostre quantos apps são classificados como “Mature 17+”.

• Mostre o top 10 apps por número de reviews bem como o respectivo número de reviews. Ordene a lista de forma decrescente por 
número de reviews.


```python
import pandas as pd
df = pd.read_csv('googleplaystore.csv')
# df.to_excel('googleplaystore.xlsx')
```

```python
df = df.drop_duplicates(subset='')
df
```

```python
df.columns = ['app', 'category', 'rating', 'reviews', 'size', 'installs', 'type', 'price', 'contentrating','genres',
              'lastupdate','currentver', 'androidver']
```

```python
df.head(1)
```

```python
df.info()
```

```python
# Limpando a coluna Installs
df['installs'] = df['installs'].apply(lambda x:x.replace('+','')) 
df['installs'] = df['installs'].apply(lambda x:x.replace(',',''))
df = df.drop(df[df['installs']=='Free'].index)
df['installs'] = pd.to_numeric(df['installs'])
df['installs']
```

```python
# Limpando a coluna Size
df_kb = df[df['size'].str.contains("k", regex=True)].copy()
df_mb = df[df['size'].str.contains("M", regex=True)].copy()
df_mb['size'] = df_mb['size'].apply(lambda x:x.replace('M',''))
df_kb['size'] = df_mb['size'].apply(lambda x:x.replace('k',''))
df_mb['size'] = pd.to_numeric(df_mb['size'])
df_kb['size'] = pd.to_numeric(df_kb['size'])
df_mb['size'] = df_mb['size'].apply(lambda x:x*1000)
df.update(df_mb)
df.update(df_kb)
df['size']
```

```python
# Limpando coluna Price
df['price'] = df['price'].apply(lambda x:x.replace('$',''))
df['price'] = pd.to_numeric(df['price'])
```

```python
# Corrigindo o tipo das colunas Last Update e Reviews
df['reviews'] = pd.to_numeric(df['reviews'])
df['lastupdate'] = pd.to_datetime(df['lastupdate'])
```

#### • Faça um gráfico de barras contendo os top 5 apps por número de instalação.

```python
import seaborn as sns
top5_intalls = df[['app', 'installs']].copy()
top5_intalls = top5_intalls.sort_values(by=['installs'], ascending=False)
top5_intalls
```

```python
top5_intalls = top5_intalls.reset_index(drop = True)
top5_intalls = top5_intalls.iloc[0:5]
sns.barplot(data = top5_intalls, y='app', x='installs')
```

#### • Faça um gráfico de pizza (pie chart) mostrando as categorias de apps existentes no dataset de acordo com a frequência em que elas aparecem.

```python
app_categorias = df['category'].copy()
app_categorias = app_categorias.value_counts()
app_categorias = pd.DataFrame(data = app_categorias)
import matplotlib.pyplot as plt
app_categorias = app_categorias.reset_index()
app_categorias.columns = ['category', 'freq']
app_categorias
```

```python
fig = plt.pie(x = app_categorias['freq'], labels = app_categorias['category'])
```

#### • Mostre qual o app mais caro existente no dataset.

```python
app_mais_caro = df[['app', 'price']].copy()    
app_mais_caro = app_mais_caro.sort_values(by = ['price'], ascending=False)
app_mais_caro.iloc[0]
```
